```python
'''An algorithm to identify the best block on a lane based on its 
poximity to three amenities namely, store, schools and gym'''

import numpy as np
import pandas as pd
```

```python
street_details =[{'gym':True, 'school':False,'store':False},
                 {'gym':False, 'school':True,'store':False},
                 {'gym':True, 'school':False,'store':False},
                 {'gym':False, 'school':False,'store':True},
                 {'gym':False, 'school':False,'store':True},
                 {'gym':False, 'school':True,'store':False},
                 {'gym':True, 'school':False,'store':True},
                 {'gym':False, 'school':False,'store':False}]
```

```python
df = pd.DataFrame(street_details)
df[['gym_prox','school_prox','store_prox']]  = len(df) +1

for col in df[['gym','school','store']]:
    for row in range(len(df)):
        for subrow in range(len(df)):
            if df.loc[subrow,'{}'.format(col)] ==True:
                df['{}_prox'.format(col)][row] = min(df['{}_prox'.format(col)][row],abs(subrow-row))
df['distance'] = df.iloc[:,3:6].max(axis=1)   
df
```
